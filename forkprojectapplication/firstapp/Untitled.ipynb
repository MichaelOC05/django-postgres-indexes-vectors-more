```python
import os
from IPython.display import Image, display

import pandasai as pai
from pandasai_litellm.litellm import LiteLLM

CLAUDE_API_KEY = os.getenv("CLAUDE_API_KEY")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

# from documentation not from tutorial as the package has been updated
llm = LiteLLM(model="gpt-4.1-mini", api_key=OPENAI_API_KEY)

pai.config.set({"llm": llm})

df = pai.read_csv("titanic_data.csv")

df.head()
```

```python
df.chat("What sex was most likely to have survived?")
```

```python
response = df.chat("Plot the histogram of survivorship of the sexes")
display(Image(response.value))
```
